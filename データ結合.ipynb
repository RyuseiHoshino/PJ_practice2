```python
merged_df = pd.merge(orders_df, rvw_df, on='order_id', how='left')
merged_df = pd.merge(merged_df, pay_df, on='order_id', how='left')
merged_df = pd.merge(merged_df, cust_df, on='customer_id', how='left')
# 次に、注文商品(item_df)をマージします。ここからデータは注文商品単位になります。
merged_df = pd.merge(merged_df, item_df, on='order_id', how='left')
# 最後に、商品と販売者の情報をマージします。
merged_df = pd.merge(merged_df, prod_df, on='product_id', how='left')
merged_df = pd.merge(merged_df, sell_df, on='seller_id', how='left')
merged_df = pd.merge(merged_df, cate_df, on='product_category_name', how='left')

# --- 3. マージ結果の確認 ---
print("--- マージ後のデータフレーム情報 ---")
merged_df.info()

print("\n\n--- マージ後のデータフレーム（先頭5行） ---")
print(merged_df.head())
```
