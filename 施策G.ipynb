## ステップ2-1：「要注意製品リスト」の作成

分析内容： まず、Olistの収益の柱でありながら、顧客満足度が低い「要注意製品」を特定します。具体的には、商品ごとの総売上（order_itemsのprice とorder_paymentsのpayment_value を合算）と、平均レビュースコア（order_reviewsのreview_score ）を算出。「総売上が上位20%以内」にもかかわらず、「平均スコアが3.5未満」といった基準で製品を絞り込み、優先的に介入すべきリストを作成します。 使用データ： order_items, order_payments, order_reviews ステップ2-2：製品情報とレビュー評価の相関分析

```python
print("### Olistデータ分析：『要注意製品リスト』の作成 ###")

# -------------------------------------------------------------------
# 1. 商品ごとの総売上を算出
# -------------------------------------------------------------------
print("\n[1/4] 商品ごとの総売上を算出中...")
# item_dfから商品ごとの価格(price)を合計
product_revenue = item_df.groupby('product_id')['price'].sum().reset_index()
product_revenue.rename(columns={'price': 'total_revenue'}, inplace=True)
print(f"...{len(product_revenue)}件の商品について総売上を算出しました。")


# -------------------------------------------------------------------
# 2. 商品ごとの平均レビュースコアを算出
# -------------------------------------------------------------------
print("\n[2/4] 商品ごとの平均レビュースコアを算出中...")
# rvw_df(レビュー)とitem_df(注文商品)をorder_idで結合し、商品とレビューを紐付ける
order_reviews_items = pd.merge(rvw_df, item_df, on='order_id')
# 商品ごとに平均スコアを計算
product_avg_score = order_reviews_items.groupby('product_id')['review_score'].mean().reset_index()
product_avg_score.rename(columns={'review_score': 'avg_review_score'}, inplace=True)
print(f"...{len(product_avg_score)}件の商品について平均スコアを算出しました。")

# -------------------------------------------------------------------
# 3. 売上データとレビュースコアデータを結合し、「要注意製品」を特定
# -------------------------------------------------------------------
print("\n[3/4] 売上データとレビューデータを結合し、『要注意製品』を特定します...")
# 商品ごとの売上と平均スコアを結合
product_stats = pd.merge(product_revenue, product_avg_score, on='product_id')

# 売上上位20%のしきい値 (80パーセンタイル) を計算
revenue_threshold = product_stats['total_revenue'].quantile(0.80)
print(f"売上上位20%のしきい値: {revenue_threshold:.2f}")

# 平均スコア3.5未満の基準を設定
score_threshold = 3.5
print(f"平均レビュースコアのしきい値: {score_threshold}")

# 両方の条件を満たす製品をフィルタリング
problem_products = product_stats[
    (product_stats['total_revenue'] >= revenue_threshold) &
    (product_stats['avg_review_score'] < score_threshold)
].copy()

print(f"\n>>> 特定された『要注意製品』の数: {len(problem_products)}件")

# -------------------------------------------------------------------
# 4. 結果の表示（商品カテゴリ名を追加して可読性を向上）
# -------------------------------------------------------------------
if not problem_products.empty:
    print("\n[4/4] 結果に商品カテゴリ名を追加して表示します...")
    # prod_df(商品)とcate_df(カテゴリ翻訳)をマージ
    products_with_translation = pd.merge(prod_df, cate_df, on='product_category_name', how='left')

    # 要注意製品リストに商品カテゴリ情報を追加
    problem_products_details = pd.merge(
        problem_products,
        products_with_translation[['product_id', 'product_category_name_english']],
        on='product_id',
        how='left'
    )

    # 表示する列を整理して、売上順にソート
    result = problem_products_details[[
        'product_id',
        'product_category_name_english',
        'total_revenue',
        'avg_review_score'
    ]].sort_values(by='total_revenue', ascending=False).reset_index(drop=True)

    print("\n--- 『要注意製品』リスト (売上順) ---")
    pd.set_option('display.max_rows', 100) # 表示行数の設定
    print(result)
else:
    print("\n>>> 『要注意製品』に該当する製品は見つかりませんでした。")
```

### 「要注意製品」の深掘り分析プラン
この分析は、3つの異なる角度から問題にアプローチします。

**分析の角度1：顧客の声（VoC）の構造分析 ―「なぜ」不満なのか？**
顧客が残したレビューコメントは、不満の根本原因を特定するための最も貴重な情報源です。

目的： 「評価が低い」という事実から一歩進み、「具体的に何が、どのように」不満の原因となっているのかを明らかにします。
分析手法：
キーワード抽出： 「要注意製品」に寄せられたレビューコメント（order_reviews.review_comment_message）から、頻出する単語（名詞、形容詞）を抽出します。「小さい」「色が違う」「壊れていた」「遅い」など、不満の傾向を大まかに掴みます 。
トピックモデリング： さらに、自然言語処理（NLP）を用いて、コメント全体を意味の近いグループ（トピック）に自動で分類します。これにより、「サイズ・寸法に関する不満」「品質・素材に関する不満」「配送に関する不満」「商品説明との乖離」といった、より具体的な不満のカテゴリとその割合を定量的に把握できます 。
得られる示唆： 製品ごとに、「この製品の低評価の60%は『サイズが合わない』という理由に起因している」といった、具体的な改善点が見つかります。

**分析の角度2：販売者（セラー）のパフォーマンス分析 ―「誰が」問題なのか？**
製品自体の問題なのか、それとも特定の販売者のオペレーションに問題があるのかを切り分けます。

目的： 製品に紐づく低評価が、製品固有の問題によるものか、特定の販売者のパフォーマンスに起因するものかを特定します。
分析手法：
同一製品・セラー別評価の比較： 「要注意製品」のリストにある各product_idについて、それを取り扱う全販売者（sellers.seller_id）を特定します。同じ製品でも、セラーによって平均レビュースコアに差がないかを比較します 。
問題セラーの特定と横断分析： もし特定のセラーの評価だけが著しく低い場合、そのセラーが販売する他の全商品の評価や配送遅延率も分析し、セラー自身のオペレーション品質に問題がないかを検証します。
得られる示唆： 「製品Aの低評価は、どのセラーが販売しても変わらないため、製品自体の見直しが必要」「製品Bの低評価は、主にセラーXYZの遅い出荷が原因である」といった、問題の所在を明確にできます。

**分析の角度3：配送プロセスの影響分析 ―「体験全体」に問題はないか？**
製品そのものに満足していても、配送体験が悪ければ総合評価は下がります。その影響度を切り分けます。

目的： 低評価の原因として、配送体験（リードタイム、遅延）がどの程度影響しているかを定量的に評価します。
分析手法：
配送状況別のスコア比較： 「要注意製品」の注文を、「配達予定日内に到着したグループ」と「遅延したグループ」に分けます（ordersテーブルのタイムスタンプを使用）。両グループで、平均レビュースコアに有意な差があるかを比較します。
送料と満足度の関係： 「要注意製品」の送料（order_items.freight_value）が、他の同価格帯・同カテゴリの製品と比較して割高でないかを分析します。高すぎる送料が、顧客の不満に繋がり、評価を押し下げている可能性を探ります 。
得られる示唆： 「この製品群では、配送が3日遅延すると平均スコアが1.2点低下する」といった、配送品質が製品評価に与える具体的なインパクトがわかります。

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
from collections import Counter
from wordcloud import WordCloud
from google.colab import drive


# ===================================================================
# 事前準備：分析対象となる「要注意製品リスト」を作成
# ===================================================================
print("\n[事前準備] 『要注意製品リスト』を作成します...")
try:
    product_revenue = item_df.groupby('product_id')['price'].sum().reset_index().rename(columns={'price': 'total_revenue'})
    order_reviews_items = pd.merge(rvw_df, item_df, on='order_id')
    product_avg_score = order_reviews_items.groupby('product_id')['review_score'].mean().reset_index().rename(columns={'review_score': 'avg_review_score'})
    product_stats = pd.merge(product_revenue, product_avg_score, on='product_id')

    revenue_threshold = product_stats['total_revenue'].quantile(0.80)
    score_threshold = 3.5
    problem_products_df = product_stats[
        (product_stats['total_revenue'] >= revenue_threshold) &
        (product_stats['avg_review_score'] < score_threshold)
    ]

    if problem_products_df.empty:
        print("INFO: 分析対象となる「要注意製品」は見つかりませんでした。ここで処理を終了します。")
        # この後の分析は実行せずに終了
        exit()

    problem_product_ids = problem_products_df['product_id'].tolist()
    print(f"...{len(problem_product_ids)}件の要注意製品を特定しました。")

except NameError:
    print("❌ エラー: item_df や rvw_df などのデータが読み込まれていません。")
    print("このセルを実行する前に、データの読み込みを完了させてください。")
    exit()
except Exception as e:
    print(f"❌ 事前準備中に予期せぬエラーが発生しました: {e}")
    exit()

# ===================================================================
# 分析の角度1：顧客の声（VoC）の構造分析
# ===================================================================
print("\n### 分析1：顧客の声（VoC）の構造分析 ###")
problem_reviews_df = order_reviews_items[
    (order_reviews_items['product_id'].isin(problem_product_ids)) &
    (order_reviews_items['review_score'].isin([1, 2]))
]
problem_comments = problem_reviews_df['review_comment_message'].dropna().astype(str).tolist()

if not problem_comments:
    print("...要注意製品に対する低評価のレビューコメントが見つかりませんでした。VoC分析をスキップします。")
else:
    print(f"...{len(problem_comments)}件の低評価レビューを分析対象とします。")
    try:
        nlp = spacy.load('ja_ginza')
        words = []
        for comment in problem_comments:
            doc = nlp(comment)
            for token in doc:
                if token.pos_ in ['NOUN', 'ADJ', 'VERB']:
                    words.append(token.lemma_)

        text_for_wordcloud = ' '.join(words)

        if text_for_wordcloud:
            print("\n...ワードクラウドを生成します。")
            stop_words = {'こと', 'もの', 'よう', 'ため', '商品', '製品', '購入', '注文', '連絡', '対応', '利用', '思う', '感じ', '自分', '時間', '写真', 'ほう', 'の', 'ん', '人', 'いる', 'する', 'なる', 'ない', 'くる', 'いく', '言う', 'やる', 'できる', 'おく', 'つく', 'くださる'}
            font_path = '/usr/share/fonts/opentype/ipafont-gothic/ipagp.ttf'
            wordcloud = WordCloud(font_path=font_path, width=800, height=400, background_color='white', stopwords=stop_words, colormap='viridis').generate(text_for_wordcloud)
            plt.figure(figsize=(15, 7))
            plt.imshow(wordcloud, interpolation='bilinear')
            plt.axis('off')
            plt.title('「要注意製品」の低評価レビュー ワードクラウド', fontsize=20)
            plt.show()
    except NameError:
        print("❌ エラー: 'nlp'変数が定義されていません。GiNZAモデルの読み込みに失敗した可能性があります。")
    except Exception as e:
        print(f"❌ VoC分析中にエラーが発生しました: {e}")


# ===================================================================
# 分析の角度2：販売者（セラー）のパフォーマンス分析
# ===================================================================
print("\n### 分析2：販売者（セラー）のパフォーマンス分析 ###")
top_problem_product_id = problem_products_df.sort_values(by='total_revenue', ascending=False).iloc[0]['product_id']
print(f"...分析例：最も売上が高い要注意製品 ({top_problem_product_id}) を分析します。")
product_specific_reviews = order_reviews_items[order_reviews_items['product_id'] == top_problem_product_id]

if product_specific_reviews.empty:
    print("...該当製品のレビューデータが見つかりませんでした。")
else:
    seller_performance = product_specific_reviews.groupby('seller_id')['review_score'].agg(['mean', 'count']).reset_index().rename(columns={'mean': 'avg_score', 'count': 'review_count'})
    print(f"\n【製品 {top_problem_product_id} のセラー別平均スコア】")
    print(seller_performance.sort_values(by='avg_score'))


# ===================================================================
# 分析の角度3：配送プロセスの影響分析
# ===================================================================
print("\n### 分析3：配送プロセスの影響分析 ###")
problem_orders_reviews = pd.merge(problem_reviews_df[['order_id', 'review_score']], orders_df, on='order_id')
time_cols = ['order_delivered_customer_date', 'order_estimated_delivery_date']
problem_orders_reviews.dropna(subset=time_cols, inplace=True)

if problem_orders_reviews.empty:
    print("...配送遅延の分析対象となるレビューデータが見つかりませんでした。")
else:
    for col in time_cols:
      problem_orders_reviews[col] = pd.to_datetime(problem_orders_reviews[col], errors='coerce')
    problem_orders_reviews.dropna(subset=time_cols, inplace=True)

    problem_orders_reviews['is_delayed'] = (problem_orders_reviews['order_delivered_customer_date'] > problem_orders_reviews['order_estimated_delivery_date'])
    delivery_impact = problem_orders_reviews.groupby('is_delayed')['review_score'].mean()

    print("\n【要注意製品における配送状況別の平均スコア】")
    if pd.notna(delivery_impact.get(False)):
        print(f"予定通り到着: {delivery_impact.get(False):.2f} 点")
    else:
        print("予定通り到着: データなし")
    if pd.notna(delivery_impact.get(True)):
        print(f"遅延して到着: {delivery_impact.get(True):.2f} 点")
    else:
        print("遅延して到着: データなし")
    if pd.notna(delivery_impact.get(False)) and pd.notna(delivery_impact.get(True)):
        diff = delivery_impact.get(False) - delivery_impact.get(True)
        print(f"スコア差: {diff:.2f} 点")
```

分析内容： 全商品を対象に、「商品説明の文字数（products.product_description_lenght ）」や「写真の枚数（products.product_photos_qty ）」といった製品情報の充実度が、レビュースコアにどのような影響を与えているかを統計的に分析します。

得られる示唆： 「写真が5枚以上の商品は、1枚の商品に比べて平均スコアが0.4ポイント高い」といった、データに基づいた「売れる・満足される商品ページの法則」を導き出します。

```python
import japanize_matplotlib


print("### 分析ステップ2-2：製品情報とレビュー評価の相関分析 ###")

# -------------------------------------------------------------------
# 1. 分析用データの準備
# -------------------------------------------------------------------
print("\n[1/4] 分析用データを準備中...")
# レビュー、注文商品、商品の各テーブルを結合
merged_df = pd.merge(rvw_df, item_df, on='order_id')
analysis_df = pd.merge(merged_df, prod_df, on='product_id')

# 分析に必要なカラムを抽出
analysis_df = analysis_df[[
    'product_id',
    'review_score',
    'product_description_lenght',
    'product_photos_qty'
]].copy()

# 欠損値を含む行を削除
analysis_df.dropna(inplace=True)
print("...データ準備完了。")


# -------------------------------------------------------------------
# 2. 手法1：相関行列とヒートマップによる全体像の把握
# -------------------------------------------------------------------
print("\n[2/4] 手法1：相関行列とヒートマップを作成中...")

# 補足：レビュー件数が少ない商品のスコアは信頼性が低いため、
# レビュー件数を算出し、一定数以上（例：10件以上）のレビューがある商品に絞って分析
product_review_counts = analysis_df['product_id'].value_counts().reset_index()
product_review_counts.columns = ['product_id', 'review_count']
analysis_df = pd.merge(analysis_df, product_review_counts, on='product_id')

# レビューが10件以上の商品に絞り込み
reliable_analysis_df = analysis_df[analysis_df['review_count'] >= 10].copy()

# 相関分析用のカラムを選択
correlation_df = reliable_analysis_df[[
    'review_score',
    'product_description_lenght',
    'product_photos_qty',
    'review_count'
]]

# 相関行列を計算
correlation_matrix = correlation_df.corr()

# ヒートマップで可視化
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('製品情報とレビュー評価の相関ヒートマップ\n(レビュー10件以上の商品)', fontsize=16)
plt.show()

print(correlation_matrix)


# -------------------------------------------------------------------
# 3. 手法2-A：写真枚数と平均レビュースコアの関係
# -------------------------------------------------------------------
print("\n[3/4] 手法2-A：写真枚数別の平均レビュースコアを分析・可視化中...")

# 写真枚数でグループ化し、平均レビュースコアを算出
# ここでもレビュー10件以上の信頼できるデータを使用
avg_score_by_photos = reliable_analysis_df.groupby('product_photos_qty')['review_score'].mean()

# 棒グラフで可視化
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_score_by_photos.index, y=avg_score_by_photos.values, palette='viridis')
plt.title('写真の枚数と平均レビュースコアの関係', fontsize=16)
plt.xlabel('写真の枚数 (枚)', fontsize=12)
plt.ylabel('平均レビュースコア (点)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(avg_score_by_photos)


# -------------------------------------------------------------------
# 4. 手法2-B：商品説明の文字数と平均レビュースコアの関係
# -------------------------------------------------------------------
print("\n[4/4] 手法2-B：商品説明文字数別の平均レビュースコアを分析・可視化中...")

# 商品説明の文字数をカテゴリに分類（ビン化）
# レビュー10件以上の信頼できるデータを使用
bins = [0, 500, 1500, reliable_analysis_df['product_description_lenght'].max()]
labels = ['短 (0-500字)', '中 (501-1500字)', '長 (1501字以上)']
reliable_analysis_df['description_len_group'] = pd.cut(
    reliable_analysis_df['product_description_lenght'],
    bins=bins,
    labels=labels,
    right=False
)

# 文字数グループで集計し、平均レビュースコアを算出
avg_score_by_desc = reliable_analysis_df.groupby('description_len_group')['review_score'].mean()


# 棒グラフで可視化
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_score_by_desc.index, y=avg_score_by_desc.values, palette='plasma')
plt.title('商品説明の文字数と平均レビュースコアの関係', fontsize=16)
plt.xlabel('商品説明の文字数カテゴリ', fontsize=12)
plt.ylabel('平均レビュースコア (点)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(avg_score_by_desc)
```
